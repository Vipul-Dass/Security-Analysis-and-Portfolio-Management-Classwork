Download the daily last 6 month closing prices for the common stock of your assigned company and calculate daily returns, expected daily return (assuming equal probability for observed historical data) and risk (standard deviation) for this expected return

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
raw = pd.read_csv('501904063_Dabur.csv')

raw.head()
```

```python
dropp = ['Unnamed: 5','Unnamed: 6','Unnamed: 7','Unnamed: 8']
data = raw.drop(dropp,axis =1)
data.head()
```

```python
data['Return'] = (data['Close Price'] - data['Prev Close'])/data['Prev Close']

data.head()
```

```python
data.describe()
```

```python
data['Probability'] = 1/124

data.describe()
```

```python
E_Ret = np.dot(data['Return'],data['Probability'])

E_Ret_P = "{:.5%}".format(E_Ret)

avg_ret = "{:.5%}".format(data['Return'].mean())

std_ret = "{:.5%}".format(data['Return'].std())



print("Expected Return assuming equal probability of the 124 observed historical daily returns:", E_Ret_P,"\n\n"
     ,"Average Return:", avg_ret,"\n\n"
     ,'Standard Deviation of this daily return:', std_ret,"\n\n"
     ,'Coefficient of Variation:', data['Return'].std()/data['Return'].mean())
```
